```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('samples_g5_ctrl.csv')
df2 = pd.read_csv('samples_g5_MG.csv')
df3 = pd.read_csv('samples_g25_ctrl.csv')
df4 = pd.read_csv('samples_g25_MG.csv')
```

```python
## g5 ##
```

```python
column_name = 'SUCD2_u6m'
sns.histplot(data=df1, x=column_name, color='blue', label='g5_ctrl')
sns.histplot(data=df2, x=column_name, color='orange', label='g5_MG')

median_df1 = df1[column_name].median()
median_df2 = df2[column_name].median()
plt.axvline(median_df1, color='blue', linestyle='dashed', linewidth=1)
plt.axvline(median_df2, color='orange', linestyle='dashed', linewidth=1)
plt.text(median_df1, plt.ylim()[1]*0.95, f'Median : {median_df1:.2f}', color='blue', ha='right')
plt.text(median_df2, plt.ylim()[1]*0.95, f'Median : {median_df2:.2f}', color='orange', ha='left')

plt.xlabel('Flux distribution in mmol/g.dw/hr')
plt.ylabel('Frequency')
plt.title('Histogram of SUCD2_u6m')
#plt.legend(loc='upper right', bbox_to_anchor=(1.02, 1))
#plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.legend()
plt.show()
```
