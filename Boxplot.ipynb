```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('samples_g5_ctrl.csv')
df2 = pd.read_csv('samples_g5_MG.csv')
df3 = pd.read_csv('samples_g25_ctrl.csv')
df4 = pd.read_csv('samples_g25_MG.csv')
```

```python
dfs = [df1, df2, df3, df4]

medians = []
std_devs = []

data_to_plot = []
for df in dfs:
    column_name = 'CER1_26ter'
    if column_name in df.columns:
        column2_data = df[column_name].fillna(0) 
        median = column2_data.median()
        std_dev = column2_data.std()
    else:
        median = 0
        std_dev = 0
    medians.append(median)
    std_devs.append(std_dev)
    data_to_plot.append(column2_data)

# Plotting boxplot
plt.figure(figsize=(8,6))
plt.boxplot(data_to_plot, labels=['g5_ctrl', 'g5_MG', 'g25_ctrl', 'g25_MG'])
plt.xlabel('Groups')
plt.ylabel('Flux in mmol/g.dw/hr')
plt.title('Boxplot of CER1_26ter')

# Adding median and std deviation annotations
for i, (median, std_dev) in enumerate(zip(medians, std_devs)):
    plt.text(i + 1, median, f'Median: {median:.2f}\nStd Dev: {std_dev:.2f}', horizontalalignment='center', verticalalignment='bottom')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
